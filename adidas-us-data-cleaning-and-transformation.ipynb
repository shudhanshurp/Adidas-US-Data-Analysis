```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('/kaggle/input/adidas-sales-dataset/Adidas US Sales Datasets.xlsx')
df.head()
```

```python
df.columns = df.iloc[3]
df.drop(range(0, 4), inplace=True)
df.dropna(axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
unq_chck_cols = ['Retailer', 'Region', 'State', 'City', 'Product', 'Sales Method']

for col in unq_chck_cols:
    print(f"Unique values in '{col}':")
    print(df[col].unique())
    print()
```

```python
df[['Price per Unit', 'Units Sold', 'Total Sales']] = df[['Price per Unit', 'Units Sold', 'Total Sales']].astype('int64')
df[['Operating Profit', 'Operating Margin']] = df[['Operating Profit', 'Operating Margin']].astype('float64')
```

```python
df['Invoice Date']=pd.to_datetime(df['Invoice Date'])
```

```python
df['Gender'] = df['Product'].apply(lambda x: 'Male' if x[0] == 'M' else 'Female')
```

```python
df["Operating Costs"] = df["Total Sales"] - df["Operating Profit"]
```

```python
df['Year'] = df['Invoice Date'].dt.year
df['Month'] = df['Invoice Date'].dt.month
df['Day'] = df['Invoice Date'].dt.day
```

```python
df.to_excel('Adidas_US_Cleaned_Data.xlsx', index=False)
```
